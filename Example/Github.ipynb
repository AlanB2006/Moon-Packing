## 📦 1. Required Software, Libraries, and Versions

<div class="alert alert-block alert-info">

This notebook relies on the REBOUND N-body integrator and its extension library software
REBOUNDx, along with standard scientific Python libraries.
Note: All simulations and data are processed through Google Colab.

</div>

### Simulation Libraries
- **REBOUND** ≥ 4.5.0  
  High-precision N-body integrator used for orbital dynamics simulations.

- **REBOUNDx** ≥ 4.6.0  
  Extension package for REBOUND that enables additional physics such as tidal forces,
  spin evolution, and dissipative effects.

### Scientific Stack
- **Python** ≥ 3.10  
- **NumPy** — numerical arrays and fast vectorized operations  
- **SciPy** — physical constants and scientific utilities  
- **Matplotlib** — data visualization and plotting  

### Utilities
- **multiprocessing** — parallel execution of simulations  
- **threading** — lightweight concurrency  
- **time / os / sys** — system interaction and timing utilities
