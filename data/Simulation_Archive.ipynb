# Simulation Archive Data

This notebook provides access to the simulation archive data used to generate the heat maps, plots, and related analysis figures in this repository.

Because the full simulation archives are too large to store directly in the GitHub repository, the data are hosted externally in [Zenodo](https://doi.org/10.5281/zenodo.20408204).


## Data Description

The Zenodo repository contains the simulation archive files used to produce the heat maps and figures shown in the paper/repository. These files include the numerical simulation outputs needed to reproduce or inspect the plotted results.
